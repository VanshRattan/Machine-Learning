```python
import numpy as np
```

```python
#Q1:Reverse the numpy array
arr=np.array([1,2,3,4,5,6])
print(arr)
print(np.flip(arr))
```

```python
#Q2: Flatten the numpy array
arr_1=np.array([[1,2,3],[2,3,4],[1,2,5]])
print(arr_1)
print("Flatten Array By Default", arr_1.flatten())  # By default it is 'C' means the row major order
print("Flatten Array By Value F", arr_1.flatten('F')) #here F tells to   flatten the array by the column major order 
print("Flatten Array By Reshape Function", arr_1.reshape(-1)) #this reshapes the numpy array  by default it is in row major order
print("Flatten Array By Reshape Function", arr_1.reshape(-1, order='F')) #this reshapes in the column major order
```

```python
#Q3: to compare the array 
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])
print(np.array_equal(arr1, arr2))  #here np.array_equal is used for comparing the array 

```

```python
#Q4:find the most frequent value and their indices  in the following array 
#counting sort also do the same function as bincount

x = np.array([1,2,3,4,5,1,2,1,1,1])  
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
x1=np.bincount(x).argmax() #this bincount tells the number of each digit in this and the argmax stores the maximum of this
y1=np.bincount(y).argmax()
print('The most frequent array in x is',x1)
print('The most frequent array in y is',y1)
print('This is the output with the indices in this:')
for i in range(len(x)):
    if(x[i]==x1):
        print(i)
    
#This is the output for the Y ARRAY
print("Most frequent value in Array Y is", y1)
print("With Indices:")
for i in range(len(y)):
    if(x[i]==x1):
        print(i)
    

```

```python
#Q5: Different sum operation
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
#This is for the sum of all the elements
print('Sum of all the elements:',np.sum(gfg))
#This is for the sum of all the elemets row wise.
print('Sum of all the elements Row Wise:',np.sum(gfg,axis=1)) #here axis=1 helps in summing the values row wise
#This is for the sum of all the elements column wise.
print('Sum of all the elements Row Wise:',np.sum(gfg,axis=0)) #here axis=0 helps in summing the values column wise
```

```python
#Q6: Different mathematical operations:
n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])
#sum of diagonal
print("Sum of diagonal elements:", np.trace(n_array)) #np.trance is used to find the sum of the diagonal elements
eigenvalues, eigenvectors = np.linalg.eig(n_array) #this stores the eigenvalue and vector  for this array given above.
#Eigen value of this: 
print("Eigenvalues:", eigenvalues)
#Eigen vector of this:
print("Eigenvectors:", eigenvectors)
#Inverse of this matrix
print("Inverse of a matrix:", np.linalg.inv(n_array))
#Determinant of this matrix
print("Determinant of matrix:", np.linalg.det(n_array))





```

```python
#Q7: Doing multiplication and finding the covariance of the following matrix
#i:
print('This is the solution for the first one:')
p = np.array([[1, 2], [2, 3]])
q = np.array([[4, 5], [6, 7]])
print("Multiplication:")
print(np.multiply(p, q)) #np.multiply is used
print("Covariance:")
print(np.cov(p, q))  #np.cov is used
#ii:
print('This is the function for second one:')
p1 = np.array([[1, 2], [2, 3], [4, 5]])
q1= np.array([[4, 5, 1], [6, 7, 2]])
print("Multiplication:")
print(np.multiply(p1, q1.T))
print("Covariance:")
print(np.cov(p1, q1))

```
