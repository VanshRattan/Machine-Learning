```python
 # Importing the Libraries
 import numpy as np
 import pandas as pd
 import matplotlib.pyplot as plt
 # Loading the Dataset
 dataset = pd.read_csv(r'E:\3rd Year\Machine Learning\Assignment\Dataset\StipendData.csv')
 X = dataset.iloc[:, :- 1].values
 y = dataset.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =1/3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)
```

```python
y_pred = linear_regressor.predict(X_test)
```

```python
# Visualizing the Training set results
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, linear_regressor.predict(X_train), color = 'blue')
plt.title('Visualization of Training Data')
plt.xlabel('Years of Research Experience')
plt.ylabel('Stipend')
plt.show()
```

```python
# Visualizing the Test set results
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, linear_regressor.predict(X_test), color = 'blue')
plt.title('Visualization of Test Data')
plt.xlabel('Years of Experience')
plt.ylabel('Stipend')
plt.show()
```

```python
#Mean absolute error
from sklearn import metrics
print(metrics.mean_absolute_error(y_test,y_pred))
```

```python
#Mean Squared Error
print(metrics.mean_squared_error(y_test,y_pred)) 
```

```python
import numpy as np
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
