Determining the datatype of each of the feature

```python
df.dtypes
```

```python
ordinal = []
continous = []
categorical_string = []

for col, dtype in df.dtypes.items():
    if dtype==int:
        ordinal.append(col)
    elif dtype==float:
        continous.append(col)
    else:
        categorical_string.append(col)
        
print("Ordinal (rank based) features : \n{o}\n\n".format(o=ordinal))
print("Continous features : \n{c}\n\n".format(c=continous))
print("Categorical string features : \n{s}\n\n".format(s=categorical_string))
```

Categorising the feature according to their data type

```python
df['Gender'].unique()
```

```python
df['Gender'] = df['Gender'].apply(lambda x : 1 if x=='M' else 0)
```

```python
df['MaritalStatus'].unique()
```

```python
df['MaritalStatus'] = df['MaritalStatus'].apply(lambda x : 1 if x=='M' else 0)
```

```python
df['LastUpdated'].unique()
```

```python
df['LastUpdated'] = df['LastUpdated'].apply(lambda x : 1 if x=='2017-03-06' else 2)
```

```python
categorical_string.remove('MaritalStatus')
categorical_string.remove('LastUpdated')
categorical_string.remove('Gender')
```

```python
df_encoded = pd.get_dummies(df,columns=categorical_string)
```

```python
df_encoded.head()
```

```python
df_encoded.dtypes.unique()
```

Applying binning methods to continuous and categorical datatype to reduce the noise

```python
df_encoded['YearlyIncome'] = pd.cut(df_encoded['YearlyIncome'],bins=5).apply(lambda x : (x.left+x.right)/2.0)
df_encoded['AvgMonthSpend'] = pd.cut(df_encoded['AvgMonthSpend'],bins=5).apply(lambda x : (x.left+x.right)/2.0)
```

Calculating similarity between yearly income and average monthly spend

```python
income_final = pd.Series(income_final.flatten())
monthly_spend_final = pd.Series(monthly_spend_final.flatten())
```

```python
#cosine similarity

def cosine_similarity(a,b):
    return np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))

CS = cosine_similarity(income_final,monthly_spend_final)
```
